### 6*. Дизамбигуация в Mystem
Одна из фичей майстема - контекстная дизамбигуация (т.е. определение правильного разбора в зависимости от контекста).

Запустите mystem на очищенном тексте с дизамбигуацией и без и проверьте есть ли какие-то различия.

```python
m = Mystem(disambiguation=False)
lemmas = m.lemmatize(clean_text)
print(''.join(lemmas))
print(len(lemmas))
```

```python
m = Mystem(disambiguation=True)
dis_lemmas = m.lemmatize(clean_text)
print(''.join(dis_lemmas))
print(len(dis_lemmas))
```

```python
for i in range(len(lemmas)):
    if lemmas[i] != dis_lemmas[i]:
        print('non dis: ' + lemmas[i - 2] + ' ' + lemmas[i] + ' ' + lemmas[i + 2])
        print('    dis: ' + dis_lemmas[i - 2] + ' ' + dis_lemmas[i] + ' ' + dis_lemmas[i + 2])
        print('-------------------------')
```

Разница анализа с дизамбигуацией и без присутствует. На мой взгляд, с параметром дизамбигуацией mystem чаще всего верно определяет леммы слов. Например, взять предложение "Всех с прошедшим Новым годом!" и слово "прошедшим". С параметром mystem определил "прошедший", что является верным разбором, без параметра лемма имеет вид "прошедшее". Также, например, с дизамбигуацией указательное местоимение "этот" правильно разбирается, в случае без дизамбигации местоимение разбирается как "это".
